## تصور العلاقة بين العوامل الديموغرافية الرقمية والتحصيل العلمي

### Subtask:
إنشاء رسوم بيانية للتشتت (Scatter Plots) أو رسوم بيانية مشتركة (Joint Plots) لاستكشاف العلاقة بين المتغيرات الديموغرافية الرقمية (مثل العمر، مستوى تعليم الوالدين) والدرجة النهائية (G3).


**Reasoning**:
To visualize the relationship between numerical demographic factors and the final grade (G3), I will import `matplotlib.pyplot` and `seaborn`, then iterate through each numerical demographic factor to create and display scatter plots.



```python
import matplotlib.pyplot as plt
import seaborn as sns

for factor in numerical_demographic_factors:
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=df, x=factor, y='G3')
    plt.title(f'Relationship between {factor.replace("_", " ").title()} and Final Grade (G3)')
    plt.xlabel(factor.replace("_", " ").title())
    plt.ylabel('Final Grade (G3)')
    plt.grid(True)
    plt.show()
```

## حساب مصفوفة الارتباط

### Subtask:
حساب مصفوفة الارتباط بين العوامل الديموغرافية الرقمية وجميع الدرجات (G1، G2، G3) لتحديد قوة واتجاه العلاقات.


**Reasoning**:
To calculate the correlation matrix, I will first combine the lists of numerical demographic factors and academic grades. Then, I will select these columns from the DataFrame, compute their correlations, and display the result using a seaborn heatmap for better visualization.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Combine the numerical_demographic_factors and academic_grades lists
all_numerical_cols = numerical_demographic_factors + academic_grades

# 2. Select these columns from the DataFrame df
correlation_df = df[all_numerical_cols]

# 3. Calculate the correlation matrix
correlation_matrix = correlation_df.corr()

# 4. Display the resulting correlation matrix using a heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Matrix of Numerical Demographic Factors and Academic Grades')
plt.show()

print("Correlation matrix calculated and visualized.")
```

## تصور العلاقة بين العوامل الديموغرافية الفئوية والتحصيل العلمي

### Subtask:
إنشاء رسوم بيانية (مثل Box Plots أو Bar Plots) لتصور كيفية تأثير العوامل الديموغرافية الفئوية (مثل الجنس، العنوان، وظيفة الأم والأب) على الدرجة النهائية (G3)، مع توفير وسيلة إيضاح.


**Reasoning**:
To visualize the impact of categorical demographic factors on the final grade (G3), I will iterate through each factor in the `categorical_demographic_factors` list and generate a box plot for each, displaying the distribution of 'G3' across different categories of the factor. This directly addresses the subtask's visualization requirement.



```python
for factor in categorical_demographic_factors:
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=df, x=factor, y='G3')
    plt.title(f'Impact of {factor.replace("_", " ").title()} on Final Grade (G3)')
    plt.xlabel(factor.replace("_", " ").title())
    plt.ylabel('Final Grade (G3)')
    plt.grid(axis='y', linestyle='--')
    plt.show()
```
